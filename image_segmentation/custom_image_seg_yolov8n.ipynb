Testing the model by making predictions from unseen test set

```python
from ultralytics import YOLO

# Load the model
model = YOLO('runs/segment/train/weights/best.pt')

# Set the source directory for your test data
source = 'test_data'

# Specify the save directory
save_dir = 'runs/segment'

# Make predictions
results = model.predict(source, save=True, project=save_dir, name='', conf=0.5)
```

```python
import glob
from IPython.display import Image, display

for img_path in glob.glob(f'runs/segment/predict/*.jpg'):
    display(Image(filename=img_path, height=600))
    print("\n")
```

```python
import cv2
import numpy as np
import torch
import os

# Create output folder
os.makedirs("masks_output", exist_ok=True)

# Iterate over all images in results
for i, result in enumerate(results):
    if result.masks is not None:  # Check if masks exist
        mask_tensor = result.masks.data  # Get the mask tensor

        for j in range(mask_tensor.shape[0]):  # Iterate over detected masks
            mask_array = mask_tensor[j].cpu().numpy()  # Convert to NumPy
            mask = (mask_array * 255).astype('uint8')  # Convert to 8-bit format
            mask = mask.squeeze()  # Ensure correct shape

            # Save the mask
            filename = os.path.join("masks_output", f"image_{i}_mask_{j}.png")
            cv2.imwrite(filename, mask)  # Save using OpenCV
            print(f"Saved: {filename}")
    else:
        print(f"No masks found for image {i}")
```
